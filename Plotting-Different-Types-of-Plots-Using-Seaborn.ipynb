## 1.0.1 Seaborn

Seaborn is one of the python data visualisation libraries that is used for plotting statistical graphics. It is built as an add-on of the matplotlib to provide more attractive plots As compared to matplotlib, seaborn is easy to tweak and has good interface for modifying and enhancing the look of Matplotlib plots. Seaborn provides better theme and styling on the top of the matplotlib.!

## 1.0.4 Bar Plot:

A bar can be plotted vertically or horizontally

# 1.0.5 Vertical bar plot:

Using the titanic dataset, we have the plot shown here. This plot shows, based on the category ‘embark_town’, the count of the titanic crash’s survivors.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_context("paper")

#load dataset
titanic= sns.load_dataset("titanic")

#create plot
sns.barplot(x="embark_town",y="age",data=titanic,palette="PuRd", errorbar=None)
print(titanic.columns)
plt.show()
```

## 1.0.6 Horizontal barplot

This plot shows the same result that is titanic crash’s survivors based on ‘embark_town’ category but with horizontal orientation.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("paper")

sns.barplot(x="age", y="embark_town",data=titanic,palette="PuRd",orient="h")
```

## 1.0.7 Histogram

With the help of histogram, using ‘iris’ dataset, ‘petal_length’ in different bins are classified and shown the number of observations in each bin using bar.

```python
import seaborn as sns
from matplotlib import pyplot as plt
df= sns.load_dataset("iris")
sns.displot(df["petal_length"],kde=False)
plt.savefig("Histogram.eps",format="eps",
            dpi=1200,bbox_inches="tight",transparent=True)
plt.show()
```

## 1.0.8 Count plot

Count plot is a histogram across a categorical variable. By considering the ‘titanic’ dataset, the plot represents using the variable ‘who’ that is what type of survivors such that (man, woman or child) falls in ‘class’ category.

```python
import matplotlib.pyplot as plt
import seaborn as sns
titanic=sns.load_dataset("titanic")
sns.countplot(x="class",hue="who",data=titanic,palette="magma")
plt.title("Survivors")
```

## 1.0.9 Line plot

A line is plotted using ‘tips’ dataset that represents tip increases as the size of the members increases

```python
tips=sns.load_dataset("tips")
sns.lineplot(data=tips,x="size",y="tip",errorbar=None)
print(tips.columns)
```

## 1.0.10 Scatter Plot

It is used to represent the relationship between two data columns. in this plot, it is observed that tip amount is less when total bill amount is less.

```python
sns.scatterplot(data=tips, x="total_bill",y="tip")
```

Further, by adding time parameter, we can visualize tips are given more during dinner time than lunch time.

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="time")
```

Now, let us see how we can style a simple Matplotlib plot using Seaborn’s set() function. For, this we have loaded the ‘tips’ into a df dataframe.

```python
from matplotlib import pyplot as plt
import seaborn as sns
df= sns.load_dataset("tips")
print(df.head())
plt.scatter(df.total_bill, df.tip)
```

Now let us see how we can style this plot using the set() function

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.scatter(df.total_bill, df.tip)
sns,set()
```

```python
# to have white grid in the background
from matplotlib import pyplot as plt
import seaborn as sns
plt.scatter(df.total_bill,df.tip)

sns.set_style("whitegrid")
plt.savefig("plot3.eps",format="eps",dpi=1200,
            bbox_inches="tight",transparent=True)
plt.show()
```

## 1.0.11 KDE plot

KDE plot is termed as Kernal Density Estimate plot. Using ‘iris’ dataset, this plot represents probability density of ‘setosa’ and ’ virginica’.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
iris=sns.load_dataset("iris")
sns.kdeplot(iris.loc[(iris["species"]=="setosa"),
                     "sepal_length"],color="b", fill=True, label="satosa")
sns.kdeplot(iris.loc[(iris["species"]=="virginica"),
                    "sepal_length"],color="r",fill=True,label="virginica")
plt.legend()
```

## 1.0.12 Box Plot

We used ‘tips’ data and boxplot() function where x-axis is representing day and y-axis is representing total bill amount. This plot represents total bill amount; on different days such as Thursday, Friday, Saturday and Sunday and few outliers can also be observed. Here, it can be observed that mean and deviation of total bill are higher on Sunday as compared to other days.

```python
import seaborn as sns
tips= sns.load_dataset("tips")
sns.boxplot(x="day",y="total_bill",data=tips)
```

## 1.0.13 Violin Plot

Violin plot is similar to box plot. In this plot density of total bill of tips dataset is represented.

```python
import seaborn as sns
tips=sns.load_dataset("tips")
ax=sns.violinplot(x=tips["total_bill"])
```

## 1.0.14 Heatmap

It is a two-dimensional plot. Values are represented in matrix form with different colors.

Using ‘flight’, the values are represented in matrix form with different colors. Here, Yellow color represents the count of passenger less than 200 and blue shows more than 600. Here it can be noticed that color gets darker as the value increases.

```python
import seaborn as sns
flights=sns.load_dataset("flights")
flights= flights.pivot("month","year","passengers")
print(flights)
```

```python
sns.heatmap(flights,linewidths=.5,cmap="YlGnBu")
```

## 1.0.15 Pair Plot

Using ‘iris’ dataset, this plot shows univariate distribution of sepal length, sepal width, petal length and petal width on different species. Also, their relationship from each other can also be visualised.

```python
import seaborn as sns
from matplotlib import pyplot as plt
df= sns.load_dataset("iris")
sns.set_style("ticks")
sns.pairplot(df,hue="species",diag_kind="kde",kind="scatter",palette="husl")
```
